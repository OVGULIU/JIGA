# BSpline Implementation in Julia

This document illustrates the functioning of the BSpline basis implementation in Julia. To use the bspline implementation, we must first include the script file and invoke the bspline module in the script file.

```julia
include("src/bspline.jl")
using bspline
```

The bspline module exports a function $N(i,p,\Xi,\xi)$ where the argument list of the function describes the $i$<sup>th</sup> basis of order $p$ corresponding to the parametric knot vector $\Xi$ evaluated at the parametric co-ordinate $\xi$. Note that the structure of the knot vector $\Xi$ must be compatible with the other arguments (i.e. $i,p,\xi$), and future implementations will check for this error.

Let us consider an open knot vector $\Xi = \{ 0,0,0,1,2,3,4,4,5,5,5 \}$. Since the first and last knot have been repeated $3$ times, and the basis is interpolatory here (the knot vector is open), we can obtain upto quadratic (order 2) basis functions.

```julia
using Gadfly
using Color

x = linspace(0,5,1000)
n1 = 9
Ξ = Float64[0.,0.,0.,1.,2.,3.,4.,4.,5.,5.,5.]
N1(i,p) = Float64[ N(i,p,Ξ,x1)  for x1 in x ]

colors = distinguishable_colors(30)
order = 2

plot([layer(x = x, y = N1(j,order), Geom.line, Theme(default_color = colors[j+2])) for j in 1:n1-order+1]...)
```

```julia
bspline.N(1,2,Ξ,0.0)
```
